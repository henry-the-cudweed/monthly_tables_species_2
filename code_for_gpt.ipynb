```python
import pandas as pd

def print_table_data(file_path):
    df = pd.read_csv(file_path)
    for index, row in df.head(1).iterrows():
        print_row(row)

def print_row(row_data):
    for key, value in row_data.items():
        print(f"{key}: {value}")
    print()

# Replace 'file_path' with the path to your file
file_path = 'data inputs\pafs_cal_ipc_invasiveness_scoring_data.csv'
print_table_data(file_path)
```
